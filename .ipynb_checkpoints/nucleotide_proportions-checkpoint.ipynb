```python
import pyfastx

# Load the FASTA file
fa = pyfastx.Fasta('rubescens_transcriptome_ORF_ignore_nested.fasta')

# Get the composition dictionary
composition = fa.composition

# Total length of the FASTA file
total_length = 176447259

# Calculate the proportion for each nucleotide
proportions = {nucleotide: round(count / total_length, 2) for nucleotide, count in composition.items()}

# Print the proportions
for nucleotide, proportion in proportions.items():
    print(f"Proportion of '{nucleotide}': {proportion}")
```

```python
import matplotlib.pyplot as plt

# Upstream Data
nucleotides = ['A', 'C', 'G', 'T']
all_mismatches = [0.30859, 0.20920, 0.21358, 0.26862] # values acquired from OL_editing.qmd
ag_mismatches = [0.33120, 0.22884, 0.21385, 0.22611]
differential_ag_mismatches = [0.35782, 0.21089, 0.16543, 0.26586]

# Plotting
bar_width = 0.3
index = range(len(nucleotides))

plt.bar(index, all_mismatches, bar_width, label='All mismatches', color='green')
plt.bar([i + bar_width for i in index], ag_mismatches, bar_width, label='AG mismatches', color='blue')
plt.bar([i + 2*bar_width for i in index], differential_ag_mismatches, bar_width, label='Differential AG mismatches', color='orange')

# Heights for dashed lines (example)
dashed_line_heights = [0.3, 0.2, 0.19, 0.31]

# Set y-axis limits
plt.ylim(0, 0.55)

# Add dashed lines with individual heights
for i, nucleotide in enumerate(nucleotides):
    plt.plot([i-bar_width/2, i+5*bar_width/2], [dashed_line_heights[i], dashed_line_heights[i]], color='purple', linestyle='dashed')

# Customize plot
plt.xlabel('Upstream nucleotide')
plt.ylabel('Proportion')
plt.xticks([i + bar_width for i in index], nucleotides)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()

```
